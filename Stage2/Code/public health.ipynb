### Visualization of BMI and Weight

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load NHANES dataset (replace 'nhanes_data.csv' with actual file path)
df = pd.read_csv('nhanes.csv')

# Add Weight in Pounds Column
df['Weight_pounds'] = df['Weight'] * 2.2

# Visualization: Histograms for BMI, Weight, Weight in Pounds, and Age
cols = ['BMI', 'Weight', 'Weight_pounds', 'Age']
df[cols].hist(bins=30, figsize=(10, 6))
plt.suptitle("Distribution of BMI, Weight, Weight in Pounds, and Age")
plt.show()
```

### Mean 60-second pulse and Range of Diastolic BP

```python
# Mean 60-Second Pulse Rate
mean_pulse = df['Pulse'].mean()
print(f"Mean 60-Second Pulse Rate: {mean_pulse:.5f}")

# Range of Diastolic Blood Pressure
min_bp = df['BPDia'].min()
max_bp = df['BPDia'].max()
print(f"Range of Diastolic Blood Pressure: {min_bp} - {max_bp}")
```

### Variance and Standard deviation

```python
# Variance and Standard Deviation of Income
income_var = df['Income'].var()
income_std = df['Income'].std()
print(f"Income Variance: {income_var:.5f}, Income Standard Deviation: {income_std:.5f}")
```
